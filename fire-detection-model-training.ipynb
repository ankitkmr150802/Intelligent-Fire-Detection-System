# Batch-4 training

```python
import os

yaml_content = """
train: /kaggle/input/chunk-datasets/Batch_4/Batch_4/images
val: /kaggle/input/chunk-datasets/valid/valid/images
test: /kaggle/input/chunk-datasets/test/test/images

nc: 1
names: ['fire']
"""

with open('/kaggle/working/data_batch4.yaml', 'w') as f:
    f.write(yaml_content)

print("‚úÖ Success: Batch 4 YAML created with exact paths!")
```

```python
from ultralytics import YOLO

# 1. Load your Phase 3 brain (Dhyan se path dekho sidebar wala)
model = YOLO('/kaggle/input/phase3/best_phase3.pt')

# 2. Deep Fine-Tuning for Batch 4
model.train(
    data='/kaggle/working/data_batch4.yaml',
    epochs=100,        # Zyada epochs accuracy ko "polish" karne ke liye
    imgsz=640,
    batch=16,
    name='fire_v11_phase4',
    lr0=0.001,         # Slow learning rate (taaki purana result ruin na ho)
    lrf=0.01,          # Final learning rate factor
    # --- Advanced Augmentations for Precision ---
    hsv_v=0.4,         # Brightness adjustment (Sunlight handle karne ke liye)
    mosaic=0.7,        # Thoda kam mosaic taaki model clear flames pe focus kare
    fliplr=0.5,        # Horizontal flips for variety
    overlap_mask=True, # Accurate bounding box positioning
    patience=25        # Agar result improve nahi hua toh 25 epochs baad auto-stop
)
```
